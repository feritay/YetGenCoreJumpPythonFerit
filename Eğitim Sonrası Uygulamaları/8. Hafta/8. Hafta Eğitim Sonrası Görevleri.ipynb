# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
 
def vkiHesaplama(boy, kilo):
    return round(boy / ((kilo/100)**2), 1)

n = int(input("Kaç giriş yapacaksiniz?"))

data = []

for i in range(n):
    adSoyad = input("Adiniz Soyadiniz:")
    yas = int(input("Yaşiniz:"))
    boy = float(input("Boy (cm):"))
    kilo = float(input("Kilo (kg):"))

    vki = vkiHesaplama(boy,kilo)

    if vki < 18.5:
        sinif = "Zayif"
    elif vki < 25:
        sinif = "Normal kilolu"
    elif vki < 30:
        sinif = "Fazla kilolu"
    elif vki < 40:
        sinif = "Obez"
    else:
        sinif = "İleri derecede obez (morbid obez)"

    kullaniciData = [adSoyad, yas, boy, kilo, vki, sinif]
    data.append(kullaniciData)

with open("veriler.txt", "w", encoding="utf-8") as file:
    file.write("Ad Soyad\tYaş\tBoy\tKilo\tVKİ\tKategori\n")
    for x in data:
        file.write("\t".join([str(i) for i in x]) + "\n")

print("Veriler başariyla kaydedildi.")
```

**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import time 

menu = {
    "ARMUT":2,
    "BÖĞÜRTLEN":5,
    "ÇİLEK":3,
    "ELMA":4,
    "KARPUZ":6,
    "KİVİ":2,
    "KİRAZ":4,
    "MUZ":4,
    "ÜZÜM":6,
}

def display_menu():
    print("----------YETBARA HOŞGELDİNİZ---------")
    print("---------------MENU-------------------")
    
    for i in list(menu.keys()):
        print(f"{i.upper()}:{menu[i]} TL")

def tekUrunFiyatHesaplama(urun, adet):
    assert adet >=0 , "Ürün adedi 0 veya 0'dan küçük olmaz!"

    if urun not in list(menu.keys()):
        return None
    else:
        return menu[urun]*adet
    
def satinAlma():
    display_menu()
    print("Lütfen satin almak istediğiniz ürünü giriniz. Çikiş yapmak için '0'i tuşlayiniz.")

    satinAlinanUrunler = []
    satinAlmaAdedi = []
    sonFiyat = 0

    while 1:
        eklenecekUrun = input("Hangi üründen eklemek istiyorsunuz?")
        if eklenecekUrun == "0":
            print("Fiş kaydediliyor.")
            break

        eklemeAdedi = int(input("Kaç adet eklemek istiyorsunuz?"))
        fiyat = tekUrunFiyatHesaplama(eklenecekUrun, eklemeAdedi)

        if fiyat is None:
            print("Yanliş bir giriş yaptiniz, ürün eklenemedi!")
        else:
            print(f"Satin alma ücreti: {fiyat} TL.\nSatin almayi onaylamak için 'y', iptal etmek için 'n' giriniz.")
            check = input("Satin almayi onayliyor musunuz?")

            if check == "y":
                print("Satin alim onaylandi.")
                print("-"*40)

                sonFiyat += fiyat

                satinAlinanUrunler.append(eklenecekUrun)
                satinAlmaAdedi.append(eklemeAdedi)

                print(f"Ara Toplam: {sonFiyat} TL")
            elif check == "n":
                print("Satin alma işlemi iptal edildi.")
                print("-"*40)

    return sonFiyat, satinAlinanUrunler, satinAlmaAdedi

def main():
    print("YetBar'a Hoş Geldiniz!\nMenüyü görüntülemek için '1'e, smoothie yapmak için '2'e basiniz. Çikmak için '0'a basiniz.")

    program = None

    while 1:
        program = int(input("Tercihiniz:"))

        if program < 0 or program > 2:
            print("Yanliş bir tercih yaptiniz. Lütfen tekrar deneyiniz.")
        elif program == 0:
            print("Çikis yapiliyor...\nSağlikli günler dileriz.")
            break
        elif program == 1:
            display_menu()
        elif program == 2:
            satinAlmaUcreti, satinAlinanUrunler, satinAlmaAdedi = satinAlma()

            with open("kasa.txt","a",encoding="utf-8") as file:
                file.write(str(time.ctime(time.time())))
                file.write("\n")
                file.write(','.join(map(str,satinAlinanUrunler)))
                file.write("\n")
                file.write(','.join(map(str,satinAlmaAdedi)))
                file.write("\n")
                file.write(f"Toplam: {satinAlmaUcreti}")
                file.write("\n")
                file.write("*"*20)
                file.write("\n")
            
            cikisKontrol = input("Başka bir işlem yapmak istiyor musunuz? (evet-hayir):")

            if cikisKontrol == "evet":
                pass
            else:
                print("\nTEŞEKKÜRLER!\nYİNE BEKLERİZ!")
                break
main()
```

```python
class Araç:
    def __init__(self, plaka, araç_türü, marka, model):
        self.plaka = plaka
        self.araç_türü = araç_türü
        self.marka = marka
        self.model = model

class Otopark:
    def __init__(self, ad, kapasite, ucret):
        self.ad = ad
        self.kapasite = kapasite
        self.ucret = ucret
        self.dolu_yer_sayisi = 0

    def yer_ekle(self):
        if self.dolu_yer_sayisi < self.kapasite:
            self.dolu_yer_sayisi += 1
            print("Yeni araç otoparka eklendi.")
        else:
            print("Otopark dolu, yer yok.")

    def yer_bosalt(self):
        if self.dolu_yer_sayisi > 0:
            self.dolu_yer_sayisi -= 1
            print("Araç otoparktan çıkarıldı.")
        else:
            print("Otopark zaten boş.")

    def bos_yer_sayisi(self):
        bos_yer = self.kapasite - self.dolu_yer_sayisi
        return bos_yer

    def ucret_hesapla(self, sure, arac_tipi,otopark_secim):
        ucret = self.ucret * sure

        if arac_tipi == "elektrikli":
         ucret *= 0.9  # %10 indirim
        elif arac_tipi == "dizel" and otopark_secim == "Yakın":
         ucret *= 0.85  # %15 indirim
        elif arac_tipi == "benzinli" and otopark_secim == "Yakın":
         ucret *= 0.8  # %20 indirim
        elif arac_tipi == "dizel" and otopark_secim == "Uzak":
         ucret *= 1  # indirim yok
        elif arac_tipi == "benzinli" and otopark_secim == "Uzak":
         ucret *= 1 # indirim yok

        return ucret

otopark1 = Otopark(ad="Kadıköy Otoparkı", kapasite=150, ucret=10)
otopark2 = Otopark(ad="Çekmeköy Otoparkı", kapasite=100, ucret=10)

def Uygulamayi_calistir():
    while True:
        print("\nYapmak istediğiniz işlemi seçin:")
        print("1. Aracınızı Sisteme Tanıtma")
        print("2. Otopark Hakkında Bilgiler")
        print("3. Bana En Yakın Otoparkı Bulma")
        print("Q. Uygulamayı Sonlandır")

        secim = input("Seçiminizi yapın: ")

        if secim == "1":
            plaka = input("Araç plakasını girin: ")
            arac_türü = input("Araç türünü girin (Elektrikli/Benzinli/Dizel): ")
            marka = input("Araç markasını girin: ")
            model = input("Araç modelini girin: ")
            arac = Araç(plaka, arac_türü, marka, model)
            print(f"{arac.plaka} plakalı aracınız, {arac.marka} {arac.model} modeli olarak sisteme tanıtıldı.")
            tercih = input("Ana menüye dönmek için 4'e, uygulamayı sonlandırmak için Q'ya basınız: ")
            if tercih == "4":
                continue
            elif tercih.upper() == "Q":
                break

        elif secim == "2":
            otopark_secim = input("Hangi otopark hakkında bilgi almak istersiniz (Kadıköy/Çekmeköy): ")
            if otopark_secim.upper() == "KADIKÖY":
                print("Otopark Adı:", otopark1.ad)
                print("Kapasite:", otopark1.kapasite)
                print("Ücret:", otopark1.ucret)
            elif otopark_secim.upper() == "ÇEKMEKÖY":
                print("Otopark Adı:", otopark2.ad)
                print("Kapasite:", otopark2.kapasite)
                print("Ücret:", otopark2.ucret)
            tercih = input("Ana menüye dönmek için 4'e, uygulamayı sonlandırmak için Q'ya basınız: ")
            if tercih == "4":
                continue
            elif tercih.upper() == "Q":
                break

        elif secim == "3":
            while True:

                ilceler = ["Şile", "Beykoz", "Çekmeköy", "Sancaktepe", "Pendik", "Tuzla", "Sultanbeyli", "Kartal", "Maltepe", "Ataşehir", "Ümraniye", "Üsküdar", "Kadıköy"]

                ilce = input("Bulunduğunuz ilçeyi girin: ")
                if ilce in ilceler:
                    if ilce in ["Kadıköy", "Üsküdar", "Ümraniye", "Ataşehir", "Maltepe", "Kartal"]:
                        print("Kadıköy'deki otopark size daha yakın ama isterseniz Çekmeköy'deki otoparka da gidebilirsiniz. Kadıköy'deki otoparkı tercih etmeniz durumunda daha az emisyon harcayacağınız için otopark indirimi alacaksınız.")
                        otopark1.yer_ekle()  # Yeni araç eklendi
                        print("Dolu Yer Sayısı:", otopark1.dolu_yer_sayisi)
                        print("Boş Yer Sayısı:", otopark1.bos_yer_sayisi())
                        arac_tipi = input("Aracınızın yakıt tipini girin (elektrikli/benzinli/dizel): ")
                        otopark_secim = input("Hangi otoparka gitmek istersiniz? (Yakın / Uzak): ")

                        if otopark_secim == "Yakın":
                            print("Kadıköy'deki otoparka yönlendiriliyorsunuz.")
                            sure = int(input("Aracın otoparkta kalma süresini girin (saat cinsinden): "))
                            ucret = otopark1.ucret_hesapla(sure, arac_tipi,otopark_secim)
                            print("Ödemeniz gereken ücret: ", ucret)
                            break

                        elif otopark_secim == "Uzak":
                            print("Çekmeköy'deki otoparka yönlendiriliyorsunuz.")
                            otopark2.yer_ekle()  # Yeni araç eklendi
                            print("Dolu Yer Sayısı:", otopark2.dolu_yer_sayisi)
                            print("Boş Yer Sayısı:", otopark2.bos_yer_sayisi())
                            sure = int(input("Aracın otoparkta kalma süresini girin (saat cinsinden): "))
                            ucret = otopark2.ucret_hesapla(sure, arac_tipi,otopark_secim)
                            print("Ödemeniz gereken ücret: ", ucret)
                            break

                        else:
                            print("Geçersiz otopark seçimi.")
                            

                    elif ilce in ["Beykoz", "Çekmeköy", "Sancaktepe", "Sultanbeyli", "Pendik", "Tuzla", "Şile"]:
                        print("Çekmeköy'deki otopark size daha yakın ama isterseniz Kadıköy'deki otoparka da gidebilirsiniz. Çekmeköy'deki otoparkı tercih etmeniz durumunda daha az emisyon harcayacağınız için otopark indirimi alacaksınız.")
                        otopark2.yer_ekle()  # Yeni araç eklendi
                        print("Dolu Yer Sayısı:", otopark2.dolu_yer_sayisi)
                        print("Boş Yer Sayısı:", otopark2.bos_yer_sayisi())
                        arac_tipi = input("Aracınızın yakıt tipini girin (elektrikli/benzinli/dizel): ")
                        otopark_secim = input("Hangi otoparka gitmek istersiniz? (Uzak / Yakın): ")

                        if otopark_secim == "Uzak":
                            print("Kadıköy'deki otoparka yönlendiriliyorsunuz.")
                            otopark1.yer_ekle()  # Yeni araç eklendi
                            print("Dolu Yer Sayısı:", otopark1.dolu_yer_sayisi)
                            print("Boş Yyer Sayısı:", otopark1.bos_yer_sayisi())
                            sure = int(input("Aracın otoparkta kalma süresini girin (saat cinsinden): "))
                            ucret = otopark1.ucret_hesapla(sure, arac_tipi,otopark_secim)
                            print("Ödemeniz gereken ücret: ", ucret)
                            break

                        elif otopark_secim == "Yakın":
                            print("Çekmeköy'deki otoparka yönlendiriliyorsunuz.")
                            otopark2.yer_ekle()  # Yeni araç eklendi
                            print("Dolu Yer Sayısı:", otopark2.dolu_yer_sayisi)
                            print("Boş Yer Sayısı:", otopark2.bos_yer_sayisi())
                            sure = int(input("Aracın otoparkta kalma süresini girin (saat cinsinden): "))
                            ucret = otopark2.ucret_hesapla(sure, arac_tipi,otopark_secim)
                            print("Ödemeniz gereken ücret: ", ucret)
                            break

                        else:
                            print("Geçersiz otopark seçimi.")

                    else:
                        print("Geçersiz ilçe.")


                tercih = input("Ana menüye dönmek için 4'e, uygulamayı sonlandırmak için Q'ya basınız: ")
                if tercih == "4":
                    break
                elif tercih.upper() == "Q":
                    break

        elif secim.upper() == "Q":
            print("Uygulama Sonlandırılıyor...İyi yolculuklar dileriz.")
            break

        else:
            print("Geçersiz bir işlem yaptınız. Lütfen tekrar deneyin.")

Uygulamayi_calistir()
```
